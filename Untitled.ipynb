```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
logs_dir = 'dump_logs'
file_names = {'max_k_loss': [file_name for file_name in os.listdir(logs_dir) 
                             if file_name.startswith('max_k_loss') 
                             and file_name.endswith('.txt')],
             'min_k_loss': [file_name for file_name in os.listdir(logs_dir) 
                             if file_name.startswith('min_k_loss') 
                             and file_name.endswith('.txt')]}
```

```python
means = [] 
for idx, file_name in enumerate(file_names['max_k_loss']):
    contents = []
    with open(os.path.join(logs_dir, file_name), 'r') as file_handler:
        contents.extend(file_handler.read().split('\n'))
    contents = np.array(contents[:-1], dtype=float)
    means.append(np.mean(contents))
x, y = zip(*[(x[0][0], x[0][1]) for x in list(zip(enumerate(means)))])
```

```python
plt.plot(x, y)
plt.xlabel('Epoch')
plt.ylabel('Avg. loss')
plt.show()
```
